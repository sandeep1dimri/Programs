# # Convert JSON file to CSV file

```python
import pandas as pd
import json
```

```python
# Open the file and load into data
with open('result2.json') as file:
    data = json.load(file)
```

```python
# Show data sample, mainly one row and more that one row
counter=0
for dt_e in data:
    counter+=1
    if counter < 25:
        continue
    print("counter:",counter," data: ",dt_e)
    if counter > 25:
        break
```

# Algorithm:
1. Read JSON file
2. Create csv in format: # frame_id, filename, class_id, name, center_x, center_y, width, height, confidence
    

```python
with open('result2.csv','w') as csv_file:
    csv_detl_head=''
    csv_detl_obj=''
    for each_entry in data:
        csv_detl_head=''
        csv_detl_obj=''
        csv_detl_head=csv_detl_head+str(each_entry.get("frame_id",'')) + ","
        csv_detl_head=csv_detl_head+each_entry.get("filename",'') + ","
        
        for each_key in each_entry:

            if "objects" in each_key and len(each_entry[each_key]): # handle object array
                for i in range (0,len(each_entry[each_key])):
                    temp_sub_dict=each_entry[each_key][i]
                    csv_detl_obj=( str(temp_sub_dict["class_id"]) + "," + temp_sub_dict["name"] + "," +
                          str(temp_sub_dict["relative_coordinates"]["center_x"]) + "," +
                          str(temp_sub_dict["relative_coordinates"]["center_y"]) + "," +
                          str(temp_sub_dict["relative_coordinates"]["width"]) + "," +
                          str(temp_sub_dict["relative_coordinates"]["height"]) + "," +
                          str(temp_sub_dict.get("confidence",''))
                         )
                    csv_file.write(csv_detl_head+csv_detl_obj+"\n")
            elif ("objects" in each_key):
                csv_file.write(csv_detl_head+ ",,,,,,"+"\n")
                
                        
                
        
```

# Result verification aka Testing

```python
# When read the output csv, add header and keep spaces as spaces
col_names=['frame_id', 'filename', 'class_id', 'name', 'center_x', 'center_y', 'width', 'height', 'confidence']
df=pd.read_csv("result2.csv",names=col_names,header=None,index_col=None)
df.fillna('', inplace=True)
```
